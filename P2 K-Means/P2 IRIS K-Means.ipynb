### Preparing the data and runing the model
First we get only the values of our 4 features, wthout column names.

```python
features = iris.iloc[:,[1,2,3,4]].values
```

##### Let's find the optimum number of clusters k 

```python
# Finding the optimum number of clusters for k-means classification
#x = iris_df.values

x = features

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.figure(figsize=(10, 6)) #dpi=80

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

As we can see the optimum number of clusters is 3, let's fit the model.

```python
kmeans = KMeans(n_clusters = 3, init = 'random',
                max_iter = 300, n_init = 10, tol=1e-04, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
plt.figure(figsize=(10, 6)) #dpi=80

plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s = 80, color='skyblue', label='Iris-setosa')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s = 80, label='Iris-versicolor')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s = 80, color='lime', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 300, c = 'red', label = 'Centroids', marker='*')

plt.xlabel("Sepal lenght", fontsize=13.5)
plt.ylabel("Sepal width", fontsize=13.5)
plt.title("Iris Clusters", fontsize=17, fontweight="bold")

plt.legend()
plt.show()
```
